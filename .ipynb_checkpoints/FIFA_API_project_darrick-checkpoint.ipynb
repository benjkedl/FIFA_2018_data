```python
hteam = []
ateam = []
h_pass_acc = []
a_pass_acc = []
h_ball_rec = []
a_ball_rec = []
winner = match_df["winner"]

for i in range(len(match_df["home_team_statistics"])):

    #grab pass accuracy and balls recovered
    hteam1 = match_df["home_team_statistics"][i]['country']
    ateam1 = match_df["away_team_statistics"][i]['country']
    hpa = match_df["home_team_statistics"][i]['pass_accuracy']
    apa = match_df["away_team_statistics"][i]['pass_accuracy']
    hbr = match_df["home_team_statistics"][i]['balls_recovered']
    abr = match_df["away_team_statistics"][i]['balls_recovered']
    hteam.append(hteam1)
    ateam.append(ateam1)
    h_pass_acc.append(hpa)
    a_pass_acc.append(apa)
    h_ball_rec.append(hbr)
    a_ball_rec.append(abr)

effort_df = pd.DataFrame({"Team":hteam + ateam,
                          "Pass Accuracy":h_pass_acc + a_pass_acc,
                          "Balls Recovered":h_ball_rec + a_ball_rec
                                     })
effort_df
```

```python
teamgamecount = effort_df["Team"].value_counts()
teamgamecount
```

```python
team_effort_df = effort_df.groupby("Team", as_index = True).sum()

team_effort_df["Number of Games"] = teamgamecount

team_effort_df.reset_index(inplace = True)

team_effort_df["Average Pass Accuracy"] = round(team_effort_df["Pass Accuracy"]/team_effort_df["Number of Games"])
team_effort_df["Average Balls Recovered"] = round(team_effort_df["Balls Recovered"]/team_effort_df["Number of Games"])

team_effort_df
```

```python
plt.bar(team_effort_df["Team"], team_effort_df["Average Pass Accuracy"], align = 'center')
plt.xticks(np.arange(32),team_effort_df["Team"], rotation = 90)
```
